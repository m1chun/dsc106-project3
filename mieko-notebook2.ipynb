```python
import pandas as pd

# Replace with your actual file paths
file1 = "mieko_data/J2_VIIRS_C2_Australia_NewZealand_VJ214IMGTDL_NRT_2024017.txt"
file2 = "mieko_data/J2_VIIRS_C2_Australia_NewZealand_VJ214IMGTDL_NRT_2024018.txt"
file3 = "mieko_data/J2_VIIRS_C2_Australia_NewZealand_VJ214IMGTDL_NRT_2024019.txt"

# Load a sample to inspect
df1 = pd.read_csv(file1)
df1.head()
```

```python
files = [file1, file2, file3]
```

```python
# Load all files into a list of DataFrames
dfs = [pd.read_csv(f) for f in files]

# Inspect one to make sure it loaded correctly
dfs[0].head()
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df['datetime'] = pd.to_datetime(df['acq_date'] + ' ' + df['acq_time'])
```

```python
df = df.drop(columns=['acq_date', 'acq_time'])
```

```python
df = df[['latitude', 'longitude', 'bright_ti4', 'frp', 'confidence', 'datetime']]
```

```python
# Save as CSV
df.to_csv("wildfires_combined.csv", index=False)

# Save as JSON
df.to_json("wildfires_combined.json", orient='records', date_format='iso')
```
